```python
df = pd.read_excel('tab-a-4.xls', skiprows=range(8))
df.columns = ['HH Type','Total HH','Non-Movers','Total Movers',
              'Total Movers from USA', 'Same County',
              'Total from Different County','Same State',
              'Different State','From Abroad']
df.head()
```

```python
# Find a row were a new dataset starts:
df.loc[df['HH Type'].str.contains('percent', case=False).fillna(False)]
```

```python
#Subset only the first dataframe:
df = df.iloc[:172,:]
```

```python
# Extract years and drop unneccessary rows:
df.loc[:,'Year'] = np.where(np.isnan(df['Total HH']), df['HH Type'],np.nan)
df.loc[:,'Year'] = df.loc[:,'Year'].ffill()
df.dropna(axis=0, subset=['Total HH'], inplace=True)
df.head()
```

```python
# Remove total rows:
df = df.loc[~df['HH Type'].str.contains('Total', case=False)]
df.head()
```

```python
# Remove total colums (leave only "Total HH"):
df.drop(columns=['Total Movers','Total Movers from USA',
                 'Total from Different County'], inplace=True)
df.head()
```

```python
# Clean HH Type column:
df.loc[:, 'HH Type'] = np.where(df['HH Type'].str.contains('owner'),'Owned','Rented')
df.head()
```

```python
# Clean Year:
df.loc[:,'Year'] = df.Year.str[:4]
df.head()
```
