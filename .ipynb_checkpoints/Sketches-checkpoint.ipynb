```python
import json

with open('data/NYC_census_tract_2010_median_income.geojson') as geofile:
    geodata = json.load(geofile)

print('There are {} entities in the geodata file'.format(len(geodata['features'])))
```

```python
sub_df = NYC.sample(n=300, random_state=0)
sub_df.head()
```

```python
sub_geodata = {"type":"FeatureCollection", "features":[]}
sub_id = sub_df['ID'].values
sub_rows = []

for i in range(len(geodata['features'])):
    sub_dict = geodata['features'][i]
    
    if sub_dict['properties']['GEOID10'] in sub_id:
        sub_rows.append(sub_dict['properties']['GEOID10'])
        sub_geodata['features'].append({'type':'Feature',
                                        'properties':{'ID':sub_dict['properties']['GEOID10']},
                                        'geometry':sub_dict['geometry']
                                        })
```

```python
print('There were {} records with geoshape data'.format(len(sub_rows)))
```

##### In the geodata the coordinates are saved in reverse order, so we need to correct it, also we don't need the third axis

```python
sub_df = sub_df.set_index('ID').loc[sub_rows].reset_index()
sub_df.shape
```

```python

venues_total_number  = NYC_venues.groupby(['ID','Category']).Subcategory.count()
venues_unique_subcat = NYC_venues.groupby(['ID','Category'], as_index=False).Subcategory.apply(lambda x: len(set(x)))
venues_integrated_count = venues_total_number * venues_unique_subcat
venues_integrated_count = venues_integrated_count.to_frame().rename(columns={0:'Score'}).reset_index()
venues_integrated_count.head()
```

```python
venues_pivot = pd.pivot_table(venues_integrated_count, index='ID', columns='Category', fill_value=0)
venues_pivot.columns = venues_pivot.columns.droplevel(0)
venues_pivot.columns.name = None
venues_pivot.head()
```

```python
venues_pivot /= venues_pivot.max(axis=0)
venues_pivot.reset_index(inplace=True)
venues_pivot.head()
```

```python
combined_df = pd.merge(left=sub_df[['ID','Median HH Income','Median HH Value']],
                       right=venues_pivot,
                       on='ID')
combined_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

np.random.seed(0)

X_train, X_test, y_train, y_test = train_test_split(combined_df.iloc[:,3:], combined_df.iloc[:,2], test_size=0.25)

lr = LinearRegression()
lr.fit(X_train, y_train)
```
