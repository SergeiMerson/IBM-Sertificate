## Part 3.2: Geodata - Processing, Transformation and Plotting<a id="3.2"></a>

##### Load geodata from geojson file that we've saved earlier:

```python
import json

with open('data/NYC_census_tract_2010_median_income.geojson') as geofile:
    geodata = json.load(geofile)

print('There are {} entities in the geodata file'.format(len(geodata['features'])))
```

Now we need to subset only relevant records:

```python
# Create dict to store relevant records:
sub_geodata = {"type":"FeatureCollection", "features":[]}

# Save 'ID' values that are present in our dataset:
sub_id = set(df_KC_clean.index)

# Iterate over records in the geofile:
for i in range(len(geodata['features'])):
    sub_dict = geodata['features'][i]
    
    # Check if the record has 'ID' that exist in the dataframe:
    if sub_dict['properties']['GEOID10'] in sub_id:
        
        # Save the data in the dict that we've created earlier:
        sub_geodata['features'].append({'type':'Feature',
                                        'properties':{'ID':sub_dict['properties']['GEOID10']},
                                        'geometry':sub_dict['geometry']
                                        })

print('There were {} records with geoshape data'.format(len(sub_geodata['features'])))
```

There was information about every census tract in our dataframe, excelent! Now we can plot the map:

```python
from geopy.geocoders import Nominatim

# Get coordinates to center the map:
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode('Kings County, New York City, NY')
lat = location.latitude
lng = location.longitude
print('Kings County coordinates is: {:.4f} latitude and {:.4f} longitude'.format(lat, lng))
```

```python
# Import libraries:
import folium
from folium.plugins import ScrollZoomToggler
from folium import Choropleth

'''
Possible color maps:
'BuGn', 'BuPu', 'GnBu', 'OrRd', 'PuBu', 'PuBuGn',
'PuRd', 'RdPu', 'YlGn', 'YlGnBu', 'YlOrBr','YlOrRd'
'''

# Create map object:
KC_map = folium.Map(location=[lat, lng+0.05], zoom_start=12, tiles='Stamen Toner')  # Shift the map a little bit to make the legent visible

# Add controls to the Map (to disable default zoom by scrolling):
KC_map.add_child(ScrollZoomToggler())

# Add color layer:
Choropleth(
    geo_data=sub_geodata,
    data=df_KC_clean.reset_index(),
    columns=['ID', 'Total Costs'],
    key_on='feature.properties.ID',
    fill_color='YlOrRd', 
    fill_opacity=0.7,
    threshold_scale=np.linspace(5000, 25000, 5, dtype=int),
    legend_name='Median Annual Total Cost of Living'
).add_to(KC_map)

KC_map
```

## Part 4: Recommendation System<a id="4"></a>

Our first step would be creation of the dataframe that would include the categories specified by Jack and Jessy alongside with their scores:

```python
# Create the dataframe from dict:
df_scores = pd.DataFrame.from_dict({'Caf & Restaurant'        : 6,
                                    'Infrastructure'           : 8,
                                    'Parks, landscape and POV' : 6,
                                    'Pet Care'                 : 5,
                                    'Health, Sport and Beauty' : 8,
                                    'Bars and Clubs'           :-8,
                                    'Tourism'                  :-7},
                                   orient='index',
                                   columns=['Score'])

# Show the dataframe:
df_scores
```

Next step - leave only records with relevant categories:

```python
df_candidates = df_all_venues.loc[df_all_venues.Category.isin(set(df_scores.index))]
df_candidates.head()
```

Count number of venues of each category by region ID:

```python
df_candidates = df_candidates.groupby(['ID','Category'], as_index=False).Subcategory.count()
df_candidates.head()
```

We need to transform the dataframe into the right form: Category labels as columns and region IDs as index. Istead of one-hot encoging (where we get 0/1 values), we will use pivot table method (because in our case it's reasonable to preserve the number of venues of each category):

```python
# Pivot the dataframe:
df_candidates = pd.pivot_table(data=df_candidates, index='ID', columns='Category', fill_value=0)

# Reset the column multiindex:
df_candidates.columns = df_candidates.columns.droplevel(0)

# Reorder columns according to the index of df_scores:
df_candidates = df_candidates[df_scores.index.tolist()]

# Print a few rows to check the result:
df_candidates.head()
```

Calculate weighted score of each region (Census Tract)!

```python
# Multiply two dataframes (matrix multiplication) and sort the result:
df_recommendations = df_candidates.dot(df_scores).sort_values(by='Score',ascending=False)

# Show first 20 rows:
df_recommendations.head()
```

Now when we have the scores for each region, we need to concatenate the columns with cooridnates and total cost of living:

```python
# Concatenate scores, venue counts and total cost of living into a single dataframe:
df_results = pd.concat([df_recommendations,
                        df_candidates,
                        df_all_clean.loc[df_recommendations.index, ['Total Costs','Latitude','Longitude']]],
                        axis=1,
                        sort=False)
# Reset the index name:
df_results.index.name = 'ID'

# Show 5 best and 5 worst regions:
df_results.head().append(df_results.tail())
```

We will normalize the Score values to make it more clear. We will use the scale from 0 to 100, where score 0 would mean that this location is the worst one for Jack and Jessy, and score of 100 - the most desirable place to live.

```python
# Apply normalization to 'Score' column, multiply by 100 and convert to integers:
df_results.loc[:, 'Score'] = ((df_results.Score - df_results.Score.min()) / (df_results.Score.max() - df_results.Score.min()) * 100).round(2)

# Show 5 best and 5 worst regions:
df_results.head().append(df_results.tail())
```

Last step: separate data about the current location of Jack and Jessy from data related to Kings County:

```python
# Extract information about the current location:
JJ_cur_loc_processed = df_results.loc['9005310200']
print(JJ_cur_loc_processed)

# Drop this record from the result dataframe:
df_results.drop(index='9005310200', inplace=True)

# Check:
print('\nRecord "9005310200" exists in the dataframe: {}'.format('9005310200' in df_results.index))
```

## Part 5: Results<a id="5"></a>

#### Absolute Score

In the previous section we got a dataframe with census tracts in Kings County that satisfy budget restrictions, rated according to Jack's and Jessy's preferences.  
If we want to suggest them locations with the highest Score, the only thing that we need to do is to take top N rows of the result dataframe:

```python
# Take Top 20 locations (by Score value):
df_Top20 = df_results.head(20)

# Show the result:
df_Top20
```

#### Optimal Options

Second possible criteria to select regions is to to choose locations that provide best score for a given sum of money:

```python
# Create dataframe to store the cooridnates of the optimal locations:
df_optimal_locations = pd.DataFrame(columns=df_results.columns)

# Create temp variable to store maximum Score:
max_score = 0

# Sort results by Total Costs and extract optimal values:
for ID, data in df_results.sort_values('Total Costs').iterrows():
    if int(data['Score']) > max_score:
        max_score = int(data['Score'])
        df_optimal_locations.loc[ID,:] = data

df_optimal_locations
```

Plot all existing options and emphasize optimal locations:

```python
# Create the plot:
fig, ax = plt.subplots(figsize=(15,15))

# Add scatterplot:
ax.scatter(df_results['Total Costs'],
           df_results['Score'],
           c='seagreen',
           edgecolors='darkgreen',
           s=50, 
           alpha=0.30,
           zorder=1)

# Plot the line of optimal choices:
ax.plot(df_optimal_locations['Total Costs'],
        df_optimal_locations['Score'],
        color='darkgreen',
        zorder=2)

# Plot points of optimal choices:
ax.scatter(df_optimal_locations['Total Costs'],
           df_optimal_locations['Score'],
           c='yellowgreen',
           edgecolors='darkgreen',
           s=50,
           zorder=3)




# Set axis parameters:
ax.set(xlim=(5000,25000),
       ylim=(0,100),
       xticks=range(5000,25001,5000),
       yticks=range(0,101,25),
       xlabel='Median Annual Cost of Living, USD',
       ylabel='Score')

# Set title:
ax.set_title('Recommended census tract characteristics', fontsize=20)

# Show the graph:
plt.show()
```

In the graph above points located bellow the line are possibly less attracitve.  
Why? Because for each of these options exists another one that provides higher score or/and lower annual cost.  
Why 'possibly'? Because this is only score representation of computed 'attractiveness' of each recommended location, in the real life it could be the situation that a location with a lower score would be much more interesting for Jack and Jessy.

#### Marginal Score Improvement

To help Jack and Jessy to decide what option to choose, we can calculate marginal score improvement for each optimal point, i.e. how many points will bring them each dollar spent for each next location improvement:

```python
# Divide df_optimal_locations by JJ_cur_loc_processed:
df_marginal_improvement = df_optimal_locations / JJ_cur_loc_processed

# Leave only relevant features and rename them:
df_marginal_improvement = df_marginal_improvement[['Score','Total Costs']].rename(columns={'Score':'Score Improvement',
                                                                                           'Total Costs': 'Total Costs Change'})

# Show the result:
df_marginal_improvement
```

```python
# Create the plot:
fig, ax = plt.subplots(figsize=(15,15))

# Name features:
X = df_marginal_improvement['Total Costs Change']
Y = df_marginal_improvement['Score Improvement']

# Plot the line:
ax.plot(X, Y, color='black', linewidth=2, marker='^', markersize=12)

# Add labels to the markers:
for ID, data in df_marginal_improvement.iterrows():
    ax.text(data['Total Costs Change'], data['Score Improvement']-0.2, ID, rotation=90)

# Set axis parameters:
ax.set(xlim=(0.25,0.9),
       ylim=(0,6),
       xticks=X.tolist(),
       yticks=Y.tolist(),
       xlabel='Relative Anual Living Cost Change',
       ylabel='Relative Score Improvement')
ax.tick_params(axis='both', rotation=30)

# Set Grid parameters:
plt.grid(linestyle='--', linewidth=1, color='lightgrey')

# Add filled regions:
ax.fill_betweenx(y=[0,6], x1=X.tolist()[0], x2=X.tolist()[1], color='lime', alpha=0.1)
ax.fill_betweenx(y=[0,6], x1=X.tolist()[2], x2=X.tolist()[3], color='lime', alpha=0.1)
ax.fill_betweenx(y=[0,6], x1=X.tolist()[4], x2=X.tolist()[-1], color='lime', alpha=0.1)

# Set title:
ax.set_title('Marginal Locations Score Improvement', fontsize=20)

# Show the graph:
plt.show()
```

First observation: all the proposed locations have Medium Annual Total Cost of Living below the current one of Jack and Lassy;  
Second observation: we can observe that there are three price groups, where additional expenses within a group in return greatly improve Score value, while transition from one group to another demands significant increase in annual expences but almost dosen't improve the location's Score

## Part 6: Discussion<a id="6"></a>

Some moments that we want to mention as possible topics for discussion:
* Recomendation system implemented in this project is very "simple" and could be futher improved, for exampe, by extracting the category scores from the list of neighborhoods/tracts with their ratings provided by the customer/user. In fact, it could be difficult to give some digital representation to how a person feels about every single venue type, not to mention how many such types could exist, but mach easier to evaluate different districts that the user knows.
* Another way to improve the system - add new features to describe the locations, for example traffic conditions, climate and ecology etc.
* In this project we made an assumption that the number of desirable/unwanted venues equally important as their scores, for example, that two venues with the score of 4 points have the same attractivness as one venue with the score of 8 points. Obviously in the real life the situation is slightly more complicated, so the system could be improved by introducing some kind of weight coefficients for each additional venue of the same type (for example, 0.5**(venue number - 1))
